TENSOR OPERATIONS:
1. Reshape: Reshapes the existing tensor into a matrix of any given dimension. The view function also works simlarly
Error is thrown if the reshape dimension is asking for more than the number of present elements
If the number of elements is unknown we can rearrange using -1 as the number of columns but the number of rows have to be specified and the data should be able to fit in the given dimension workspace. We can also use -1 in the rows but the number of columns have to be specified then. Atleast one parameter row or column must be a known value.
2. View: The view function works exactly like the reshape, but very minute differences are present between the 2. 
When the above functions are used the changes to the original tensor will show in the reshaped one as well so any number updated in the original tensor will show up even on the reshaped tensor.
3. Slices: Like slicing an array we can access parts of the tensor using slicing


```python
import torch
mytensor = torch.arange(1,10)
print(mytensor)
```

```python
mytensor1 = mytensor.reshape(3,3)
print(mytensor1)
```

```python
mytensor2= mytensor1[:,1]
print(mytensor2)
mytensor3= mytensor1[1,:]
print(mytensor3)
mytensor4= mytensor1[:,1:]
print(mytensor4)
mytensor5= mytensor1[1:3,0:2]
print(mytensor5)
```

TENSOR ARITHMATIC:
1. Add: Use + operator between two tensors, torch.add(t1,t2) will do the same thing. Will add each item with the corresponding item in the other tensor.
2. Suntract: Use - operator, torch.sub(t1,t2) will perform the same activity
3. Multiplication: Use * Operator, torch.mul(t1,t2) will perform the same activity
4. Division: Use / operator, THIS WILL CONVERT EACH MEMBER TO A FLOAT. torch.div(t1,t2) will perform the same activity
5. Remainder: Use % operator for the modulus, torch.remainder(t1,t2) will perform the same activity
6. Exponents: torch.pow(t1,t2), this is t1**t2
Note:
t1.add_(t2) functions as t1=t1+t2. This is reassignment

```python
t1 = torch.tensor([1,2,3])
t2 = torch.tensor([4,5,6])
t3 = t1 + t2
print(t3)   
#t4 = torch.matmul(t1.float(), t2.float())
#print(t4)       
t5 = t1 - t2
print(t5)
t6 = t1 * t2
print(t6)   
t7 = t2 / t1
print(t7)   
t8 = t1%t2
print(t8)   
t9 = torch.pow(t1,t2)
print(t9)
```
