This library allows us to plot and graph math functions and it's highly used to judge and assess ML models

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from numpy.random import randn
#Create a dataframe to work on
my_df = pd.DataFrame(randn(100,4), columns=['Mon','Tue','Wed','Thur'])
print(my_df) 
```

Now that we have created the dataframe we can move on to create the histogram plot for the same df.
1. Histograms

```python
my_df["Tue"].hist()
```

We can make this plot easier to read and more granular by designating bins that is the number of bars in the histogram

```python
my_df["Tue"].hist(bins=60)
```

.plot(kind="hist") performs the same function as the above without the grid lines anyways. We can get rid of gridlines using grid=False as well. We can also set legend=Wed

2. Area plots

```python
my_df.abs().plot(kind='area') #converts all data to positive to plot
```

In the argument for .plot() shading can be changed to make it lighter but give it a value betweeen 0, 1. Also title is another settable parameter in .plot and also add a legend if needed. we can also make it unstacked like this. 

```python
my_df['Tue'].abs().plot(kind='area')
```

```python
my_df[['Mon','Tue']].abs().plot(kind='area')
```

3. Bar charts

```python
my_df.abs().plot(kind="bar")
```

To increase it's readability we can use grid and stacked parameters

```python
my_df.abs().plot(kind="bar",stacked=True)
```

Title, legend, shading can all be varied in the parameters. Also .plot.bar() function also does the same thing. 

4. Linecharts

```python
my_df.abs().plot(kind='line')
```

We can alter the width of these lines, plot only for some columns, add title, legend and change the shading, change the figure size.
.plot.line() also works the same way as kind='line'

5. Scatter plots

```python
my_df.abs().plot(kind="scatter",x="Mon",y="Tue")
```

We can add a weightage called c variable. 

```python
my_df.abs().plot(kind="scatter",x="Mon",y="Tue", c="Wed")
```

cmap is a colormap that can be added to the scatterplot, we can also change size, title etc. 

6. Box plot

```python
my_df.plot(kind="box", grid=True)
```

All other parameters that can be varied in plot() are also applicable in the box map. The line in the middle shows median, the box is the central 50% and the lines are whiskers and the circles are outliers refer to notes for further.
IQR rule is taken down in notes already.

NOTE: LEGENDS DO NOT WORK ON BOX PLOTS
The grid is essentially useful for box plots

7. Hexbin plots

```python
my_df.plot(kind="hexbin",x="Mon",y="Tue")
```

To make it more legible and readable, we must adjust the gridsize accordingly

```python
my_df.plot(kind="hexbin",x="Mon",y="Tue",gridsize=15)
```

All other parameters adjustable under the plot() function can be executed on this. The color scal directly corresponds to intensity of the points in the bin.

8. Density and KDE plots

```python
my_df.plot.density()
```

Alpha, shading, and other components of the parameter of plot() are adjustable values. 

Scipy and seaborn.ked are other installations that can be used to better enhance the given plots. 

```python
my_df["Mon"].plot.kde()
```

```python
my_df.plot.kde()
```
