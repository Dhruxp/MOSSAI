This will showcase the use of pandas library in python and demonstrate how it works

```python
import pandas as pd
```

First we take a python list then convert it into a pandas series, this is shown below

```python
my_series = [10, 20, 30, 40, 50]
my_var = pd.Series(my_series)
print(my_var)
```

dtype is datatype and its called int64 i.e. integer 64 bit.
Operations can be done on this in this way:
1. Access a specific number in this using the index next to it.
2. Label using the indexes i.e change 0,1,2,3,4 to whatever I want, it's important that changing the label doesn't mean replacing the index if 30 is labelled "c", I can still use the index 2 to access and play with it

```python
my_index = ['a', 'b', 'c', 'd', 'e']
my_var2 = pd.Series(my_series, index=my_index)
print(my_var2)
print(my_var2['c'])
```
